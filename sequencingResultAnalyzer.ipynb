Welcome to the sequencing results analyzer! <br>
This alogrithm helps you to analyze your sequencing results from microsynth in no time. Simply copy paste the optained results from directly below and (optionally) provide the sequence of your gene that was cloned into the plasmid to check for potentional errors. In case you cloned a RNAi plasmid, you have to indicate this in the sequence from the 20/32 primer by adding "rnai" to the sequence in any place (see below). 

Runtime information: the alogrithm makes use of the needlemann-wunsch alogrithm for sequence alignments. Since I am too lazy to implement this myself, an implementation by NameNeedle is used. This package has to be installed by the google server first which is why the run will take probably like 10 seconds. May you fogive me.

```r
#@title 
#packages
require(stringr)
install.packages("NameNeedle")
require(NameNeedle)
#functions 
reverse_complement = function(seq)
{ #takes ATGC characters but also
  # R for A or G
  # Y for C or T
  # S for G or C
  # W for A or T
  # K for G or T
  # M for A or C
  # B for C, G, or T
  # D for A, G, or T
  # H for A, C, or T
  # V for A, C, or G
  # N for A, G, C, or T
  
  seq = unlist(strsplit(seq,""))
  
  #compute complement
  seq_complement = rep(NA,length(seq))
  seq_complement[seq=="A"] = "T"
  seq_complement[seq=="T"] = "A"
  seq_complement[seq=="C"] = "G"
  seq_complement[seq=="G"] = "C"
  seq_complement[seq=="R"] = "Y"
  seq_complement[seq=="Y"] = "R"
  seq_complement[seq=="S"] = "S"
  seq_complement[seq=="W"] = "W"
  seq_complement[seq=="K"] = "M"
  seq_complement[seq=="M"] = "K"
  seq_complement[seq=="B"] = "V"
  seq_complement[seq=="V"] = "B"
  seq_complement[seq=="D"] = "H"
  seq_complement[seq=="H"] = "D"
  seq_complement[seq=="N"] = "N"
  
  #compute reverse of complement
  seq_reverse_complement = rep(NA,length(seq))
  seq_reverse_complement[length(seq):1] = seq_complement[1:length(seq)]
  
  return(paste(seq_reverse_complement,sep="",collapse=""))
}


#######################################################################################################

sequencing_results_analyzer = function(seq1,seq2,target)
{ #this algorithm takes sequencing results form a forward primer (20/32 on our plasmids)
  #and a reverse primer (20/33 on our plasmids) which will overlapp in the middle.
  #the two sequences are combined to the final sequence which is encoded on the plasmid
  #and this plasmid is aligned against the gene that was cloned into the plasmid (target)
  #to check for sequence errors.

  #check if we have a RNAi plasmid
  rnai=FALSE
  if(str_detect(seq2,"rnai")){
    rnai = TRUE
    cat("Input sequences are from an RNAi plasmid and will be analyzed this way.\n")
  }

  seq_20_32 = str_replace_all(seq1,"[^ATGCNRYSWKMBDHV]","")
  seq_20_33 = str_replace_all(seq2,"[^ATGCNRYSWKMBDHV]","")

  #break is the sequences are too short for analysis
  if(nchar(seq_20_32)<101 | nchar(seq_20_33)<101) {cat("one or both provided sequecing results are too short. \n They are 20/33:\n",
                                                        seq_20_32,"\nand 20/33:\n",seq_20_33)
                                                        break()}
  
  if(!rnai)
  { #non rnai plasmid analysis

    #reverse complement of 20_33 is needed
    seq_20_33 = reverse_complement(seq_20_33)
    cat("the reverse complement of the sequencing result with primer 20/33 is: \n",seq_20_33,"\n\n")
    
    #scan for overlapp in the sequences between 20_32 and 20_33 revcomp.
    length_20_32 = nchar(seq_20_32)
    sequence_combined = NA
    for(i in seq(round(length_20_32/2),length_20_32,length.out = 20)[-20])
    { #the loop is simply to check 20 parts of the 2. half of 20_32 sequence 
      #for presence in the 20_33 revcomp sequence
      query = substr(seq_20_32,i,i+50) #value 50 randomely chosen by me
      
      if(str_detect(query,"[NRYSWKMBDHV]")){
        #if sequence has sequencing error don't use it,
        #go to next one
        sequencing_error = TRUE
        next()
      }else{sequencing_error = FALSE}
      
      if(str_detect(seq_20_33,query)){
        sequence_combined = paste(substr(seq_20_32,1,i+50),
                                  substr(seq_20_33,str_locate(seq_20_33,query)[2]+1,nchar(seq_20_33)),
                                  sep="",collapse="")
        break()
      }
    }
    if(!is.na(sequence_combined)){
      #print the combined sequence encoded on the plasmid
      cat("the sequence is: \n",sequence_combined)
      cat("\nthe sequence was generated with the first",i+50,
          "nucleotides of the 20/32 sequence \nand last",nchar(seq_20_33)-(str_locate(seq_20_33,query)[2]),
          "nucleotides of the reverse complement of the 20/33 sequence\n\n")
    }else{
      #if no combined sequence could be generated, report this together with
      #help of what might be the problem.
      if(sequencing_error){cat("no combined sequence could be generated",
                              "sequencing errors were detected which could be the problem\n\n")}
      else{cat("please check input sequences, something unexpected went wrong.\n")}
    }
    
    #analysis of the open reading frame in the combined sequence against 
    #the gene that was cloned into the plasmid. 
    
    #detect the open reading frame (OFR) and check it against a provided sequence
    if(str_detect(sequence_combined,"ATG([ATGC]{3})*((TGA)|(TAA)|(TAG))"))
    {
      #detect open reading frame (ORF)
      regex = "ATG([ATGC]{3})+(?=((TGA)|(TAA)|(TAG)))"
      ORF = sequence_combined
      while(str_detect(ORF,regex)){
        #a while loop to detect the correct open reading frame
        #starting at the first atg and ending at the closest inframe stop codon
        ORF = str_extract(ORF,regex)
        if(!str_detect(regex,"\\^")) {regex = paste0("^",regex)}
      }
      cat("an open reading frame was detected, the sequence is:\n",ORF,"\n\n")
      
      #prepare target sequence for alignment
      target = str_replace_all(target,"[^ATGCNRYSWKMBDHV]","")
      
      #change parameters for needlemann wunsch algorithm to select for long alignments
      myParam <- defaultNeedleParams
      myParam$GAPCHAR = "-"
      myParam$GAP = -5
      myParam$MATCH = 2

      alignment = needles(pattern = target,subject = ORF,params = myParam)

      #print this a little nicer into the console
      pattern = unlist(strsplit(as.character(alignment$align1),""))
      subject = unlist(strsplit(as.character(alignment$align2),""))

      matching = rep(" ",length(pattern))

      #prepare to print into console
      matching[which(pattern==subject)] = "|"
      matching = paste(matching,sep="",collapse="")

      alignment.out = as.character(c(alignment$align1, matching, alignment$align2))

      cat("the sequence alignment to the target sequence looks as follows:\n",
      "(1: target gene, 2: matching/not matching, 3: open reading frame obtained from sequencing)\n")
      print(alignment.out)

    }else{cat("No open reading frame could be detected in the sequence.")}
  }


  if(rnai)
  { #analysis of RNAi plasmids is straight forward:
    #make sequence alignment using NeedlemannWunsch alogrithm
    #of both inserts to target sequence

    #prepare target sequence for alignment
      target = str_replace_all(target,"[^ATGCNRYSWKMBDHV]","")
    #change parameters for needlemann wunsch algorithm to select for long alignments
      myParam <- defaultNeedleParams
      myParam$GAPCHAR = "-"

    #forward insert
      
      alignment1 = needles(pattern = target,subject = seq_20_32,params = myParam)

      #the algorithm introduces lots of frame shifts in the end of the pattern,
      #because this is short than the sequencing result. So we remove them by hand
      alignment1$align1 = paste(substr(alignment1$align1,1,round(nchar(alignment1$align1)/4)),
                                  str_replace_all(substr(alignment1$align1,round(nchar(alignment1$align1)/4)+1,nchar(alignment1$align1)),"-",""),
                                  sep="",collapse="") #ugly long but works


      #print this a little nicer into the console
      pattern1 = unlist(strsplit(as.character(alignment1$align1),""))
      subject1 = unlist(strsplit(as.character(alignment1$align2),""))

       #some cosmetics
      pattern1 = c(pattern1,rep("-",length(subject1)-length(pattern1)))

      matching1 = rep(" ",length(pattern1))

      #prepare to print into console
      matching1[which(pattern1==subject1)] = "|"
      matching1 = paste(matching1,sep="",collapse="")
      pattern1 = paste(pattern1,sep="",collapse="")

      alignment.out.1 = as.character(c(pattern1, matching1, alignment1$align2))

      cat("\nthe sequence alignment of the forward insert to the target sequence looks as follows:\n",
      "(1: target gene, 2: matching/not matching, 3: RNAi forward insert (primer: 20/32))\n")
      print(alignment.out.1)

    #reverse insert
      
      alignment2 = needles(pattern = target,subject = seq_20_33,params = myParam)
      #the sequence must not be the reverse complement, as the insert has to be reversed!

      #the algorithm introduces lots of frame shifts in the end of the pattern,
      #because this is short than the sequencing result. So we remove them by hand
      alignment2$align1 = paste(substr(alignment2$align1,1,round(nchar(alignment2$align1)/4)),
                                  str_replace_all(substr(alignment2$align1,round(nchar(alignment2$align1)/4)+1,nchar(alignment2$align1)),"-",""),
                                  sep="",collapse="") #ugly long but works

      #print this a little nicer into the console
      pattern2 = unlist(strsplit(as.character(alignment2$align1),""))
      subject2 = unlist(strsplit(as.character(alignment2$align2),""))

      #some cosmetics
      pattern2 = c(pattern2,rep("-",length(subject2)-length(pattern2)))

      matching2 = rep(" ",length(pattern2))

      #prepare to print into console
      matching2[which(pattern2==subject2)] = "|"
      matching2 = paste(matching2,sep="",collapse="")
      pattern2 = paste(pattern2,sep="",collapse="")

      alignment.out.2 = as.character(c(pattern2, matching2, alignment2$align2))

      cat("\n\nthe sequence alignment of the reverse insert to the target sequence looks as follows:\n",
      "(1: target gene, 2: matching/not matching, 3: RNAi reverse insert (primer: 20/33))\n")
      print(alignment.out.2)
  }


}

sequencing_results_analyzer(seq1 = sequence_20_32,seq2 = sequence_20_33,target = target_gene)

```

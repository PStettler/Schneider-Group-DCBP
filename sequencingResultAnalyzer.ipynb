Runtime information: the alogrithm makes use of the Needleman Wunsch alogrithm for sequence alignments. Since I am too lazy to implement this myself, an implementation by NameNeedle is used.
```r
#@title 
#packages
require(stringr)
install.packages("NameNeedle")
require(NameNeedle)
#functions 
reverse_complement = function(seq)
{ #takes ATGC characters but also
  # R for A or G
  # Y for C or T
  # S for G or C
  # W for A or T
  # K for G or T
  # M for A or C
  # B for C, G, or T
  # D for A, G, or T
  # H for A, C, or T
  # V for A, C, or G
  # N for A, G, C, or T
  
  seq = unlist(strsplit(seq,""))
  
  #compute complement
  seq_complement = rep(NA,length(seq))
  seq_complement[seq=="A"] = "T"
  seq_complement[seq=="T"] = "A"
  seq_complement[seq=="C"] = "G"
  seq_complement[seq=="G"] = "C"
  seq_complement[seq=="R"] = "Y"
  seq_complement[seq=="Y"] = "R"
  seq_complement[seq=="S"] = "S"
  seq_complement[seq=="W"] = "W"
  seq_complement[seq=="K"] = "M"
  seq_complement[seq=="M"] = "K"
  seq_complement[seq=="B"] = "V"
  seq_complement[seq=="V"] = "B"
  seq_complement[seq=="D"] = "H"
  seq_complement[seq=="H"] = "D"
  seq_complement[seq=="N"] = "N"
  
  #compute reverse of complement
  seq_reverse_complement = rep(NA,length(seq))
  seq_reverse_complement[length(seq):1] = seq_complement[1:length(seq)]
  
  return(paste(seq_reverse_complement,sep="",collapse=""))
}


translateDNAtoAA = function(seq)
{ #an alogirthm for (fast) DNA to AA translation
  #takes a string of ATGC DNA characters that are a
  #open reading frame (ATG...STOP) and returns a 
  #string of amino acids

  #packages
  require(stringr)
  
  #get codons and unique codons
  seq.tripplets = str_extract_all(seq,"[ATGC]{3}")[[1]]
  seq.tripplets.unique = unique(seq.tripplets)
  
  { #load codon table
    codons.1 = c(rep("T",16),rep("C",16),rep("A",16),rep("G",16))
    codons.2 = rep(c(rep("T",4),rep("C",4),rep("A",4),rep("G",4)),4)
    codons.3 = rep(c("T","C","A","G"),16)
    one.letter.not = c("F","F","L","L","S","S","S","S","Y","Y","*","*","C","C","*","W",
                       "L","L","L","L","P","P","P","P","H","H","Q","Q","R","R","R","R",
                       "I","I","I","M","T","T","T","T","N","N","K","K","S","S","R","R",
                       "V","V","V","V","A","A","A","A","D","D","E","E","G","G","G","G")
    codons = cbind(paste(codons.1,codons.2,codons.3,sep=""),one.letter.not)
  }#load codon table end


  #translation to amino acids
  aaString = rep(NA,length(seq.tripplets))
  #find aa of every unique codon and combine to aa string
  for (i in 1:length(seq.tripplets.unique)){
    aa = codons[which(seq.tripplets.unique[i]==codons[,1]),2]
    pos.of.codon = which(seq.tripplets==seq.tripplets.unique[i])
    aaString[pos.of.codon] = aa
  }
  
  return(paste(aaString,sep="",collapse=""))

}



#######################################################################################################

sequencing_results_analyzer = function(seq1,seq2,target)
{ #this algorithm takes sequencing results form a forward primer (20/32 on our plasmids)
  #and a reverse primer (20/33 on our plasmids) which will overlapp in the middle.
  #the two sequences are combined to the final sequence which is encoded on the plasmid
  #and this plasmid is aligned against the gene that was cloned into the plasmid (target)
  #to check for sequence errors.

  #check if we have an RNAi plasmid
  rnai=FALSE
  if(str_detect(seq2,"rnai")){
    rnai = TRUE
    cat("Input sequences are from an RNAi plasmid and will be analyzed this way.\n")
  }

  seq_forward_primer = str_replace_all(seq1,"[^ATGCNRYSWKMBDHV]","")
  seq_reverse_primer = str_replace_all(seq2,"[^ATGCNRYSWKMBDHV]","")

  #break is the sequences are too short for analysis
  if(nchar(seq_forward_primer)<101 | is.null(seq_forward_primer) | is.na(seq_forward_primer)) {
    cat("The provided forward primer sequencing result is too short \n The sequence is: \n",seq_forward_primer,)
    break()
    }
  
  if(!rnai)
  { #non rnai plasmid analysis

    #a revers primer result sequence is optional, check if it was provided 
    #if it was not provided, shortcut to aligments.
    #if it was provided, compute combined sequence.
    if(seq_reverse_primer=="") {sequence_combined = seq_forward_primer}
    else{
      #reverse complement of reverse primer result is needed
      seq_reverse_primer = reverse_complement(seq_reverse_primer)
      cat("the reverse complement of the sequencing result with the reverse primer is: \n",seq_reverse_primer,"\n\n")
      
      #scan for overlapp in the sequences between forward and reverse primer revcomp.
      length_forward_primer = nchar(seq_forward_primer)
      sequence_combined = NA
      for(i in seq(round(length_forward_primer/2),length_forward_primer,length.out = 20)[-20])
      { #the loop is simply to check 20 parts of the 2. half of forward primer result sequence 
        #for presence in the reverse primer result revcomp sequence
        query = substr(seq_forward_primer,i,i+30) #value 30 randomely chosen by me
        
        if(str_detect(query,"[NRYSWKMBDHV]")){
          #if sequence has sequencing error don't use it,
          #go to next one
          sequencing_error = TRUE
          next()
        }else{sequencing_error = FALSE}
        
        if(str_detect(seq_reverse_primer,query)){
          sequence_combined = paste(substr(seq_forward_primer,1,i+30),
                                    substr(seq_reverse_primer,str_locate(seq_reverse_primer,query)[2]+1,nchar(seq_reverse_primer)),
                                    sep="",collapse="")
          break()
        }
      }
    }
    if(!is.na(sequence_combined)){
      #print the combined sequence encoded on the plasmid
      cat("the sequence is: \n",sequence_combined)
      cat("\nthe sequence was generated with the first",ifelse(seq_reverse_primer=="",nchar(seq_forward_primer),i+30),
          "nucleotides of the forward primer sequencing \n")
      cat = ifelse(seq_reverse_primer=="","\n",paste("and the last",nchar(seq_reverse_primer)-(str_locate(seq_reverse_primer,query)[2]),"nucleotides of the reverse complement of the reverse primer sequencing\n\n"))
      cat(cat)
    }else{
      #if no combined sequence could be generated, report this together with
      #help of what might be the problem.
      if(sequencing_error){cat("no combined sequence could be generated",
                              "sequencing errors were detected which could be the problem\n\n")}
      else{cat("please check input sequences, something unexpected went wrong.\n")}
    }
    
    #analysis of the open reading frame in the combined sequence against 
    #the gene that was cloned into the plasmid. 
    #check if target gene was provided and prepare target sequence for alignment
      target = str_replace_all(target,"[^ATGCNRYSWKMBDHV]","")
      if(target==""){cat("\n no target gene comparison can be made as no sequence",
                          "\n was provided. Job done.");break()}
    
    #detect the open reading frame (OFR) and check it against a provided sequence
    if(str_detect(sequence_combined,"ATG([ATGC]{3})*((TGA)|(TAA)|(TAG))"))
    {
      #detect open reading frame (ORF)
      regex = "ATG([ATGC]{3})+(?=((TGA)|(TAA)|(TAG)))"
      ORF = sequence_combined
      while(str_detect(ORF,regex)){
        #a while loop to detect the correct open reading frame
        #starting at the first atg and ending at the closest inframe stop codon
        ORF = str_extract(ORF,regex)
        if(!str_detect(regex,"\\^")) {regex = paste0("^",regex)}
      }
      cat("an open reading frame was detected, the sequence is:\n",ORF,"\n\n")
      
      #change parameters for Needleman Wunsch algorithm
      myParam <- defaultNeedleParams
      myParam$GAPCHAR = "-"

      alignment = needles(pattern = target,subject = ORF,params = myParam)

      #needles from NameNeedle is bugging around in some local maxima/minima
      #especially when one seqeunce is much longer than the other one.
      #as a result it will introduce unlogical frame shifts in the end of the 
      #short sequence to expand its length. This is annoying and I cannot fix
      #it because i cannot punish many short more than one long frameshift.
      #therefor, I will fix this by removing all these downstream frame shifts
      #manually and hope that I don't remove an actual one, although this would
      #still be clearly visible in the output...
      if(nchar(target) < nchar(ORF)){
        #remove indels in pattern
        alignment$align1 = paste(substr(alignment$align1,1,str_locate(alignment$align1,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2] -1),
                                    str_replace_all(substr(alignment$align1,str_locate(alignment$align1,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2],nchar(alignment$align1)),"-",""),
                                    sep="",collapse="") #ugly af but works
                                    #".*[ATGCNRYSWKMBDHV]{30,}[-]" is a regex to find the first downstream frame shift induced by the bug.


        #print this a little nicer into the console
        pattern = unlist(strsplit(as.character(alignment$align1),""))
        subject = unlist(strsplit(as.character(alignment$align2),""))

        #some cosmetics
        pattern = c(pattern,rep("-",length(subject)-length(pattern)))
        }
      if(nchar(target) > nchar(ORF)){
        #remove indels in subject
        alignment$align2 = paste(substr(alignment$align2,1,str_locate(alignment$align2,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2] -1),
                                    str_replace_all(substr(alignment$align2,str_locate(alignment$align2,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2],nchar(alignment$align2)),"-",""),
                                    sep="",collapse="") #ugly af but works
                                    #".*[ATGCNRYSWKMBDHV]{30,}[-]" is a regex to find the first downstream frame shift induced by the bug.


        #print this a little nicer into the console
        pattern = unlist(strsplit(as.character(alignment$align1),""))
        subject = unlist(strsplit(as.character(alignment$align2),""))

        #some cosmetics
        subject = c(subject,rep("-",length(pattern)-length(subject)))
        }
      if(nchar(target) == nchar(ORF)){
        #no indels have to be removed.
        #print this a little nicer into the console
        pattern = unlist(strsplit(as.character(alignment$align1),""))
        subject = unlist(strsplit(as.character(alignment$align2),""))

        }
      

      matching = rep(" ",length(pattern))
      #prepare to print into console
      matching[which(pattern==subject)] = "|"
      matching = paste(matching,sep="",collapse="")
      pattern = paste(pattern,sep="",collapse="")

      alignment.out = as.character(c(pattern, matching, alignment$align2))

      cat("the ORF sequence alignment to the target sequence looks as follows:\n",
      "(1: target gene, 2: matching/not matching, 3: open reading frame obtained from sequencing)\n")
      l = 1
      continue.print = TRUE
      while(continue.print){
        if(l==nchar(pattern)){continue.print = FALSE}
        else{pattern.out = substr(pattern,l,l+99)
        matching.out = substr(matching,l,l+99)
        alignment.out = substr(as.character(alignment$align2),l,l+99)
        print(c(pattern.out,matching.out,alignment.out))
        l = l+100
        l = ifelse(l<=nchar(pattern),l,nchar(pattern))
        cat("\n")
        }
      }
      

      #translate ORF and target gene to protein (AA) and compare sequences
      if(!str_detect(ORF,"[^ATGC]")){
      ORF_translate = translateDNAtoAA(ORF)
      }else{cat("\nsequencing results contain unread bases and cannot be translated.\n");break()}
      cat("\n\n the translated open reading frame is:\n",ORF_translate)

      HA_tag_count = str_count(ORF_translate,"YPYDVPDYA")
      myc_tag_count = str_count(ORF_translate,"EQKLISEEDL")
      TEV_site_count = str_count(ORF_translate,"ENLYFQGS")

      cat("\n\nthe translated sequence contains",HA_tag_count,"HA tags ;",myc_tag_count,"myc tags ; ",TEV_site_count,"TEV cleavage sites.")

      #detect open reading frame in target gene and translate it (only if all bases are ATGC)
      if(str_detect(target,"ATG([ATGC]{3})*((TGA)|(TAA)|(TAG))") & !str_detect(target,"[^ATGC]"))
      {
      #detect open reading frame (ORF)
      regex = "ATG([ATGC]{3})+(?=((TGA)|(TAA)|(TAG)))"
      ORF_target = target
      while(str_detect(ORF_target,regex)){
        #a while loop to detect the correct open reading frame
        #starting at the first atg and ending at the closest inframe stop codon
        ORF_target = str_extract(ORF_target,regex)
        if(!str_detect(regex,"\\^")) {regex = paste0("^",regex)}
      }
      }else{cat("\nprovided target gene does not contain an open reading frame",
                  "\nand/or unread bases and thus cannot be translated.\n");break()}

      #translate the open reading frame of the target sequence
      ORF_target_translate = translateDNAtoAA(ORF_target)

      #compare the translated sequences using the Needleman Wunsch algorithm

      alignment_aa = needles(pattern = ORF_target_translate, subject = ORF_translate, params = myParam)

      #needles from NameNeedle is bugging around in some local maxima/minima
      #especially when one seqeunce is much longer than the other one.
      #as a result it will introduce unlogical frame shifts in the end of the 
      #short sequence to expand its length. This is annoying and I cannot fix
      #it because i cannot punish many short more than one long frameshift.
      #therefor, I will fix this by removing all these downstream frame shifts
      #manually and hope that I don't remove an actual one, although this would
      #still be clearly visible in the output...
      if(nchar(ORF_target_translate) < nchar(ORF_translate)){
        #remove indels in pattern
        alignment_aa$align1 = paste(substr(alignment_aa$align1,1,str_locate(alignment_aa$align1,".*[A-Z]{10,}[-]")[2] -1),
                                    str_replace_all(substr(alignment_aa$align1,str_locate(alignment_aa$align1,".*[A-Z]{10,}[-]")[2],nchar(alignment_aa$align1)),"-",""),
                                    sep="",collapse="") #ugly af but works
                                    #".*[A-Z]{10,}[-]" is a regex to find the first downstream frame shift induced by the bug.


        #print this a little nicer into the console
        pattern_aa = unlist(strsplit(as.character(alignment_aa$align1),""))
        subject_aa = unlist(strsplit(as.character(alignment_aa$align2),""))

        #some cosmetics
        pattern_aa = c(pattern_aa,rep("-",length(subject_aa)-length(pattern_aa)))
      }

      if(nchar(ORF_target_translate) > nchar(ORF_translate)){
        #remove indels in subject
        alignment_aa$align2 = paste(substr(alignment_aa$align2,1,str_locate(alignment_aa$align2,".*[A-Z]{10,}[-]")[2] -1),
                                    str_replace_all(substr(alignment_aa$align2,str_locate(alignment_aa$align2,".*[A-Z]{10,}[-]")[2],nchar(alignment_aa$align2)),"-",""),
                                    sep="",collapse="") #ugly af but works
                                    #".*[A-Z]{10,}[-]" is a regex to find the first downstream frame shift induced by the bug.


        #print this a little nicer into the console
        pattern_aa = unlist(strsplit(as.character(alignment_aa$align1),""))
        subject_aa = unlist(strsplit(as.character(alignment_aa$align2),""))

        #some cosmetics
        subject_aa = c(subject_aa,rep("-",length(pattern_aa)-length(subject_aa)))
      }

      if(nchar(ORF_target_translate) > nchar(ORF_translate)){
        #nothing has to be changed
        #print this a little nicer into the console
        pattern_aa = unlist(strsplit(as.character(alignment_aa$align1),""))
        subject_aa = unlist(strsplit(as.character(alignment_aa$align2),""))
      }

      matching_aa = rep(" ",length(pattern_aa))
      #prepare to print into console
      matching_aa[which(pattern_aa==subject_aa)] = "|"
      matching_aa = paste(matching_aa,sep="",collapse="")
      pattern_aa = paste(pattern_aa,sep="",collapse="")

      alignment.out_aa = as.character(c(pattern_aa, matching_aa, alignment_aa$align2))

      cat("\n\nthe translated ORF sequence alignment to the translated target sequence looks as follows:\n",
      "(1: target gene aa, 2: matching/not matching, 3: translated open reading frame obtained from sequencing)\n")

      l = 1
      continue.print = TRUE
      while(continue.print){
        if(l==nchar(pattern_aa)){continue.print = FALSE}
        else{pattern_aa.out = substr(pattern_aa,l,l+99)
        matching_aa.out = substr(matching_aa,l,l+99)
        alignment_aa.out = substr(as.character(alignment_aa$align2),l,l+99)
        print(c(pattern_aa.out,matching_aa.out,alignment_aa.out))
        l = l+100
        l = ifelse(l<=nchar(pattern_aa),l,nchar(pattern_aa))
        cat("\n")
        }
      }


    }else{cat("No open reading frame could be detected in the sequence.\n")}
  }


  if(rnai)
  { #analysis of RNAi plasmids is straight forward:
    #make sequence alignment using Needleman Wunsch alogrithm
    #of both inserts to target sequence

    #check if target gene was provided and prepare target sequence for alignment
      target = str_replace_all(target,"[^ATGCNRYSWKMBDHV]","")
      if(target==""){cat("\n no target gene comparison can be made as no sequence",
                          "\n was provided. Job done.");break()}

    #change parameters for Needleman Wunsch algorithm
      myParam <- defaultNeedleParams
      myParam$GAPCHAR = "-"

    #forward insert
      alignment1 = needles(pattern = target,subject = seq_forward_primer,params = myParam)

      #needles from NameNeedle is bugging around in some local maxima/minima
      #especially when one seqeunce is much longer than the other one.
      #as a result it will introduce unlogical frame shifts in the end of the 
      #short sequence to expand its length. This is annoying and I cannot fix
      #it because i cannot punish many short more than one long frameshift.
      #therefor, I will fix this by removing all these downstream frame shifts
      #manually and hope that I don't remove an actual one, although this would
      #still be clearly visible in the output...
      #we only do this in the pattern sequence because this (in RNAi plamids)
      #will always be a sequence <600 bp whereas sequencing reactions usually 
      #give around 1000 bp.
      alignment1$align1 = paste(substr(alignment1$align1,1,str_locate(alignment1$align1,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2] -1),
                                  str_replace_all(substr(alignment1$align1,str_locate(alignment1$align1,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2],nchar(alignment1$align1)),"-",""),
                                  sep="",collapse="") #ugly af but works
                                  #".*[ATGCNRYSWKMBDHV]{30,}[-]" is a regex to find the first downstream frame shift induced by the bug.


      #print this a little nicer into the console
      pattern1 = unlist(strsplit(as.character(alignment1$align1),""))
      subject1 = unlist(strsplit(as.character(alignment1$align2),""))

      #some cosmetics
      pattern1 = c(pattern1,rep("-",length(subject1)-length(pattern1)))

      matching1 = rep(" ",length(pattern1))

      #prepare to print into console
      matching1[which(pattern1==subject1)] = "|"
      matching1 = paste(matching1,sep="",collapse="")
      pattern1 = paste(pattern1,sep="",collapse="")

      alignment.out.1 = as.character(c(pattern1, matching1, alignment1$align2))

      cat("\nthe sequence alignment of the forward insert to the target sequence looks as follows:\n",
      "(1: target gene, 2: matching/not matching, 3: RNAi forward insert (primer: 20/32))\n")

      l = 1
      continue.print = TRUE
      while(continue.print){
        if(l==nchar(pattern1)){continue.print = FALSE}
        else{pattern1.out = substr(pattern1,l,l+99)
        matching1.out = substr(matching1,l,l+99)
        alignment1.out = substr(as.character(alignment1$align2),l,l+99)
        print(c(pattern1.out,matching1.out,alignment1.out))
        l = l+100
        l = ifelse(l<=nchar(pattern1),l,nchar(pattern1))
        cat("\n")
        }
      }


    #reverse insert
      alignment2 = needles(pattern = target,subject = seq_reverse_primer,params = myParam)
      #the sequence must not be the reverse complement, as the insert has to be reversed!

      #the algorithm introduces lots of frame shifts in the end of the pattern,
      #because this is short than the sequencing result. So we remove them by hand
      alignment2$align1 = paste(substr(alignment2$align1,1,str_locate(alignment2$align1,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2] -1),
                                  str_replace_all(substr(alignment2$align1,str_locate(alignment2$align1,".*[ATGCNRYSWKMBDHV]{30,}[-]")[2],nchar(alignment2$align1)),"-",""),
                                  sep="",collapse="") #ugly af but works
                                  #".*[ATGCNRYSWKMBDHV]{30,}[-]" is a regex to find the first downstream frame shift induced by the bug.

      #print this a little nicer into the console
      pattern2 = unlist(strsplit(as.character(alignment2$align1),""))
      subject2 = unlist(strsplit(as.character(alignment2$align2),""))

      #some cosmetics
      pattern2 = c(pattern2,rep("-",length(subject2)-length(pattern2)))

      matching2 = rep(" ",length(pattern2))

      #prepare to print into console
      matching2[which(pattern2==subject2)] = "|"
      matching2 = paste(matching2,sep="",collapse="")
      pattern2 = paste(pattern2,sep="",collapse="")

      alignment.out.2 = as.character(c(pattern2, matching2, alignment2$align2))

      cat("\n\nthe sequence alignment of the reverse insert to the target sequence looks as follows:\n",
      "(1: target gene, 2: matching/not matching, 3: RNAi reverse insert (primer: 20/33))\n")

        l = 1
      continue.print = TRUE
      while(continue.print){
        if(l==nchar(pattern2)){continue.print = FALSE}
        else{pattern2.out = substr(pattern2,l,l+99)
        matching2.out = substr(matching2,l,l+99)
        alignment2.out = substr(as.character(alignment2$align2),l,l+99)
        print(c(pattern2.out,matching2.out,alignment2.out))
        l = l+100
        l = ifelse(l<=nchar(pattern2),l,nchar(pattern2))
        cat("\n")
        }
      }
  }


}

sequencing_results_analyzer(seq1 = sequence_forward_primer,seq2 = sequence_reverse_primer,target = target_gene)

```
